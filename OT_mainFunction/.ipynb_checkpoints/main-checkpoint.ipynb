```python
# def plot_data(time_range_all, ssDNA_all_percentage, distance_range_all, bead_size, junction_position_all, basepairs):
def plot_DNA_force_data(time_from_exo, time_to_exo, time_from_pol, time_to_pol, filename, junction_position_all, ssDNA_all_percentage, cycle, bead_size, basepairs,direction='forward'):

    """
    Plots ssDNA%, end-to-end distance, DNA polymerase trace, and basepairs as subplots.
    
    Args:
        time_range_all (array-like): Time range for the data points.
        ssDNA_all_percentage (array-like): ssDNA percentages for each time point.
        distance_range_all (array-like): Distances for each time point.
        bead_size (float): Bead size.
        junction_position_all (array-like): Junction positions for each time point.
        basepairs (array-like): Basepair values for each time point.
    """ 
    # here you will select and analyze your interesting cycle with corresponding time range; 
    data_ranges = get_data_ranges(time_from_exo, time_to_exo, time_from_pol, time_to_pol, filename)
    # all time range of ROI
    time_range_all = data_ranges['time_range_all']
    distance_range_all = data_ranges['distance_range_all']

    font = {'family': 'Arial', 'weight': 'normal', 'size': 16}
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

    # Plot 1: ssDNA% as a function of time
    ax1.set_ylabel('ssDNA %', fontdict=font)
    ax1.set_xlabel('Time (s)', fontdict=font)
    ax1.scatter(time_range_all/1000, ssDNA_all_percentage * 100, color='black', linestyle='dashed', s=0.5, label='End-to-End Distance')
    ax1.set_ylim(-5, 90)

    # Plot 2: End-to-End Distance and DNA Polymerase Trace
    ax2.set_ylabel('Distance(um)', fontdict=font)
    ax2.set_xlabel('Time (s)', fontdict=font)
    ax2.scatter(time_range_all/1000, distance_range_all - bead_size, color='black', linestyle='dashed', s=0.5, label='End-to-End Distance')
    ax2.scatter(time_range_all/1000, junction_position_all, color='green', s=0.5, label='DNA Polymerase Trace')
    
    time_range_all = np.array(time_range_all, dtype=np.float64)
    distance_range_all = np.array(distance_range_all, dtype=np.float64)
    bead_size = np.array(bead_size, dtype=np.float64)
    junction_position_all = np.array(junction_position_all, dtype=np.float64)
    basepairs = np.array(basepairs, dtype=np.float64)

    if direction == 'forward': # default direction is 'forward'
        ax2.fill_between(time_range_all/1000, distance_range_all - bead_size, junction_position_all, alpha=0.5)
    elif direction == 'reverse':
        ax2.fill_between(time_range_all/1000, junction_position_all, alpha=0.5)
    else:
        raise ValueError("Invalid direction value. Please use 'forward' or 'reverse'.")

    # ax2.fill_between(time_range_all/1000, distance_range_all - bead_size, junction_position_all, alpha=0.5)
    ax2.legend()
    ax2.set_ylim(0, 4.3)
    ax2.invert_yaxis()
    ax2.xaxis.set_ticks_position('top')

    # Plot 3: Basepair changes as a function of time
    ax3.set_xlabel('Time (s)', fontdict=font)
    ax3.set_ylabel('Basepairs', fontdict=font)
    ax3.plot(time_range_all/1000, basepairs, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=2, label='Basepairs')
    ax3.legend()
    ax3.set_title('BasePair Changes as a Function of Time', fontdict=font)

    plt.tight_layout()
    plt.show()
    
    plt.savefig(filename.replace('.tdms', '-cycle#') + cycle + '-CombinedPlots'+'.svg', format='svg', dpi=300, bbox_inches='tight')
```

```python
plot_DNA_force_data(time_from_exo, time_to_exo, time_from_pol, time_to_pol, filename, junction_position_all, ssDNA_all_percentage, cycle, bead_size, basepairs,direction='forward')
```
